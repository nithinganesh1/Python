```python
#to Qury a website
import requests

link = requests.get("https://en.wikipedia.org/wiki/ICC_Men%27s_Player_Rankings").text
```

```python
print(link)
```

```python
#pulling info brom webpage
from bs4 import BeautifulSoup
soup = BeautifulSoup(link, 'lxml')
```

```python
#prettify helps arange in a structure
print(soup.prettify())
```

```python
soup.title.string
```

```python
soup.a
```

```python
all_links= soup.find_all('a')
print(all_links)
```

```python
for link in all_links:
    print(link.get('href'))
```

```python
all_tables = soup.find_all('table')
print(all_tables)
```

```python
ODI_cricket_Top_Players = soup.findAll('table',class_="wikitable")[13]
print(ODI_cricket_Top_Players)
```

```python
import pandas as pd
ODI_cricket_Top_Players = soup.find_all('table', class_='wikitable')[13]

ODI_cricket_Top_Players = pd.read_html(str(ODI_cricket_Top_Players))[0]
ODI_cricket_Top_Players
```
