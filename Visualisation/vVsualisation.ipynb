```python
from random import choice
class RandomWalk:
    """A class to generate random walk"""
    def __init__(self,num_points=5000):
        """Initialize attributes of a walk"""
        self.num_points =num_points
        
        #All walk start at (0,0)
        self.xvalues=[0]
        self.yvalues=[0]
    def fill_walk(self):
        """Calculate all points in the walk"""
        #keep walking steps until the walk reach the desired length.
        while len(self.xvalues)<self.num_points:
            #diside wich direction will go and how far go.
            x_direction = choice([1,-1])
            x_distance = choice([0,1,2,3,4])
            x_step = x_direction*x_distance
            
            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4])
            y_step = y_direction*y_distance
            
            #reject moves that go nowhere.
            if x_step == 0 and y_step == 0:
                continue
                
            #calculate the new posision
            x=self.xvalues[-1] + x_step
            y=self.yvalues[-1] + y_step
            
            self.xvalues.append(x)
            self.yvalues.append(y)
```

```python
import matplotlib.pyplot as plt

while True:
    rw=RandomWalk()
    rw.fill_walk()
    #ploting using matplotlib.pyplot

    plt.scatter(rw.xvalues,rw.yvalues,s=10,c=rw.yvalues,cmap=plt.cm.Blues,edgecolors='none')
    plt.xlabel('xvalues',c='r')
    plt.ylabel('yvalues',c='r')
    plt.title('Random Walk',fontsize=14,c='b')
    plt.show()

    a=input("Need another Random Walk y/n")
    if a=='n':
        break
```
